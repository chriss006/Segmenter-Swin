```python
import os
import cv2
import numpy as np

input_dir = './datasets/labels' # 기존 레이블 디렉토리
output_dir = './datasets/label' # 변환된 레이블 저장 디렉토리
os.makedirs(output_dir, exist_ok=True)


for filename in os.listdir(input_dir):
    if filename.endswith('.png'):
        label_path = os.path.join(input_dir, filename)
        label = cv2.imread(label_path, cv2.IMREAD_UNCHANGED)
        label[label == 128] = 1  # Convert 128 to 1
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, label)
        print(f"Processed {filename}: Unique classes {np.unique(label)}")
```

```python
label_dir = './datasets/label/'
class_counts = np.zeros(2)  # [background, meniscus]

for filename in os.listdir(label_dir):
    label_path = os.path.join(label_dir, filename)
    label = cv2.imread(label_path, cv2.IMREAD_UNCHANGED)
    unique, counts = np.unique(label, return_counts=True)
    for u, c in zip(unique, counts):
        class_counts[u] += c

total_pixels = np.sum(class_counts)
class_ratios = class_counts / total_pixels
print(f"Class ratios [background kidney]: {class_ratios}")  # 클래스별 비율 출력
```

```python
def calculate_class_ratios(txt_file, label_dir):

    # txt 파일에서 데이터 리스트 읽기
    with open(txt_file, 'r') as f:
        file_list = [line.strip() for line in f.readlines()]
    
    # 클래스 픽셀 개수 저장용
    class_counts = {}

    for file_name in file_list:
        # 각 파일의 레이블 경로
        label_path = os.path.join(label_dir, file_name + '.png') # 레이블 확장자를 적절히 변경하세요
        label = cv2.imread(label_path, cv2.IMREAD_UNCHANGED)

        if label is None:
            print(f"Failed to load label: {label_path}")
            continue

        # 각 클래스의 픽셀 개수 계산
        unique, counts = np.unique(label, return_counts=True)
        for cls, count in zip(unique, counts):
            if cls not in class_counts:
                class_counts[cls] = 0
            class_counts[cls] += count

    # 전체 픽셀 수 계산
    total_pixels = sum(class_counts.values())

    # 클래스 비율 계산
    class_ratios = {cls: count / total_pixels for cls, count in class_counts.items()}

    return class_ratios

# 경로 설정
txt_file = "../SMC/datasets/splits/val.txt"  # Validation 데이터 리스트 경로
label_dir = "../SMC/datasets/train/labels"  # 레이블 이미지 디렉토리 경로

# 클래스 비율 계산
class_ratios = calculate_class_ratios(txt_file, label_dir)
print("클래스 비율:", class_ratios)
```
