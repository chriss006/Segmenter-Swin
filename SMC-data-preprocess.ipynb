```python
import numpy as np
import cv2
import os

# 레이블 파일 경로
label_dir = '../SMC/dataset/train/labels'
```

```python
for filename in os.listdir(label_dir):
    label = cv2.imread(os.path.join(label_dir, filename), cv2.IMREAD_UNCHANGED)
    unique_values = np.unique(label)
    print(f'{filename}: {unique_values}')
```

```python
import cv2
import matplotlib.pyplot as plt

# 라벨 이미지 읽기
label_image = cv2.imread('../SMC/dataset/train/labels/a11_0141.dcm.png', cv2.IMREAD_GRAYSCALE)

# 값 확인
print("Unique values in the label image:", np.unique(label_image))

# 시각화
plt.imshow(label_image, cmap='gray')
plt.colorbar()
plt.show()
```
